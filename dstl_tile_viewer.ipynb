```python
%matplotlib inline

from tifffile import imread, imshow, imsave
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
import math
```

```python
from os import listdir
```

```python
imgs = listdir('../../datasets/DSTL/three_band/')
```

```python
for image in imgs:
    img = imread('../../datasets/DSTL/three_band/' + image)
    bitspersample = int(math.ceil(math.log(img.max(), 2)))
    img >>= bitspersample - 8
    img = img.astype('B')
    cv2.imwrite('../../datasets/DSTL/three_band/' + image[:-4] + '.png',cv2.merge([img[2],img[1],img[0]]).astype(np.uint8))
```
